# Iran's most populous cities

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import numpy as np
```

```python
metroCities = ["Tehran", "Mashhad", "Isfahan", "Karaj", "Shiraz", "Tabriz", "Qom", "Ahvaz", "Kermanshah", "Urmia"]
population = np.array([15800000, 3989070, 3416000, 3416000, 1800000, 1773023, 1260000, 1184788, 1083833, 1000000])
latitude = [35.4120, 36.1935, 32.3955, 35.4945, 29.3636, 38.0454, 34.3824, 31.1817, 34.2013, 37.3238]
longitude = [51.2323, 59.3236, 51.4013, 50.5803, 52.3233, 46.1757, 50.5235, 48.4042, 47.0528, 45.0353]
colors = ["mediumblue", "grey", "darkorchid", "darkorange", "cornflowerblue", "forestgreen", "yellowgreen", "gold", "lightpink", "crimson"]
```

```python
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.xlim([44,63])
plt.ylim([25,39])
plt.title("Iran's most populous cities")
plt.grid(True)

for i in range (10):
    plt.scatter(longitude, latitude, s=population/100000*1.5, c=colors[i] , label=metroCities[i] ,alpha=1)
    plt.legend(scatterpoints=1,title="City",loc="lower right")

plt.scatter(longitude, latitude, s=population/100000*1.5, c=colors)

plt.show()
```

```python
# create new figure, axes instances.
fig=plt.figure()
ax=fig.add_axes([0.1,0.1,0.8,0.8])
# setup mercator map projection.
m = Basemap(llcrnrlon=42.,llcrnrlat=23.,urcrnrlon=68.,urcrnrlat=42.,\
            rsphere=(6378137.00,6356752.3142),\
            resolution='l',\
            lat_0=40.,lon_0=-20.,lat_ts=20.)

m.drawcoastlines()
m.fillcontinents()

# draw parallels
m.drawparallels(np.arange(20,45,4),labels=[1,1,0,1])

# draw meridians
m.drawmeridians(np.arange(40,67,4),labels=[1,1,0,1])
ax.set_title("Iran's most populous cities")

# draw plot
for i in range (10):
    m.scatter(longitude, latitude, s=population/100000*1.5, c=colors[i] , label=metroCities[i] ,alpha=1)
    ax.legend(scatterpoints=1,title="City",loc="lower right")
    
m.scatter(longitude, latitude ,s=population/100000*1.5, c=colors)

plt.show()
```
